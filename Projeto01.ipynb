#Manipulando dados

```python
import pandas as pd

vendas = pd.read_excel('/content/drive/MyDrive/Python Data Science: Básico ao Avançado - com projetos reais/Projeto 01/Vendas/Vendas+-+Base+de+Dados.xlsx')
display(vendas)
```

**Passo 02: Calcular produto mais vendido**

```python
tb_vendas = vendas.groupby('Produto').sum()
display(tb_vendas)
```

```python
tb_vendas_02 = tb_vendas[['Quantidade', 'Valor Unitário']]
display(tb_vendas_02)
```

Colocando em ordem crescente ou descrescente uma coluna escolhida

```python
#sort_values serve ordenar os dados
 #(by -> escolher a coluna ,
 #ascending -> TRUE = Crescente)
tb_vendas_03 = tb_vendas_02.sort_values(by='Quantidade', ascending= True)
display(tb_vendas_03)
```

<!-- **negrito**Passo 02 -->

```python
#ascending -> FALSE = Decrescente
tb_vendas_03 = tb_vendas_02.sort_values(by='Quantidade', ascending= False)
display(tb_vendas_03)
```

Passo 03: Verificando mais vendido em faturamento

```python
#Coluna nova criada para realizar um dados novo
tb_faturamento = vendas
tb_faturamento['Faturamento'] = tb_faturamento['Quantidade'] * tb_faturamento['Valor Unitário']
display(tb_faturamento)
```

```python
#Separando as tabelas e colocando uma tabela com Produtos e o valor do faturamento somente
tb_faturamento = tb_faturamento.groupby('Produto').sum()
tb_faturamento_produto = tb_faturamento['Faturamento'].sort_values(ascending=False)
display(tb_faturamento_produto)
```

Passo 04:Qual foi a loja que mais vendeu?

```python
tb_loja = vendas.groupby('Loja').sum()
tb_loja = tb_loja[['Quantidade', 'Valor Unitário', 'Faturamento']].sort_values(by='Quantidade', ascending=False)
display(tb_loja)
```

Passo 05: Ticket médio por loja

Passo 06: Criar um gráfico/Dashboard da loja/estado que mais vendeu (em faturamento)

```python
vendas ['Ticket Medio'] = vendas['Valor Unitário']
#Anteriormente fizemos este agrupamento para conseguir realizar a média dos valores
#tb_ticket_medio = vendas.groupby('Loja').sum()
tb_ticket_medio = vendas.groupby('Loja').mean(numeric_only=True)
tb_ticket_medio = tb_ticket_medio[['Ticket Medio']].sort_values(by='Ticket Medio', ascending=False)
display(tb_ticket_medio)
```

```python
#Como utilizou o group by para criar a tabela da loja, teve que usar '.index' para o programa conseguir entender que o "x" seria a informação das lojas no caso o índice (em inglês - index) da tabela
import plotly.express as px
gr_loja = px.bar(tb_loja, x=tb_loja.index , y='Faturamento')
gr_loja.show()
```

```python
gr_loja = px.area(tb_loja, x=tb_loja.index , y='Faturamento')
gr_loja.show()
```

Puxando formatação para deixar no formato de modelo de dinheiro

```python
from babel.numbers import format_currency
```

```python
#Puxando a tabela com uma coluna
#'.apply' é um código de aplicação para alguma condição, no caso colocando uma função dentro da coluna
#Essa função serve para o lambda ir em cada numero da coluna e colocar o valor em padrão de moeda brasileira
vendas_formatado = tb_vendas_03['Valor Unitário'].apply(lambda x: format_currency(x, 'BRL', locale='pt_BR'))
print(vendas_formatado)
```

```python
#Colocando todos os valores no modelo Dataframe para conseguir deixar a tabela um pouco mais apresentável
#Primeiro coloca em uma sequencia decrescente
vendas_formatado = tb_vendas_03[['Valor Unitário']].sort_values(by='Valor Unitário', ascending=False)
#Segundo transforma a tabela em um DataFrame (pq ela perde o formato do pandas) e ao mesmo tempo ele aplica a função lambda
#Essa função executa cada linha da caluna até a última ajustando para a configuração de moeda
vendas_formatado = pd.DataFrame((vendas_formatado['Valor Unitário'].apply(lambda x: format_currency(x, 'BRL', locale='pt_BR'))))
display(vendas_formatado)
```

Deixando a parte numérica decimal mais visível

.replace = trocando caracter pelo outro

```python
#Sequencia decrescente
tb_decimais = tb_faturamento[['Valor Unitário', 'Faturamento']].sort_values(by='Faturamento', ascending=False)
#Colocando todo o valor com decimais
tb_decimais = tb_decimais['Valor Unitário'].apply(lambda x: ('{:,.2f}'.format(x).replace(',','.'))),tb_decimais['Faturamento'].apply(lambda y: ('{:,.2f}'.format(y).replace(',','.')))
#Transformando as tabelas em dataframe pq ela perde o formato pandas após utilizar lambda
tb_decimais = pd.DataFrame(tb_decimais)
display(tb_decimais)
```
